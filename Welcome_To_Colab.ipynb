```
import pandas as pd
```

```
import numpy as np
```

```
import seaborn as sns
```

```
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('https://raw.githubusercontent.com/rolexpaolo/FINAL-ACTIVITY2/refs/heads/main/SALARY_DATA.csv')
```

```
print(df)
```

```
salary = df['Salary'].to_numpy()
```

```
mean_salary = np.mean(salary)
```

```
print(mean_salary)
```

```
median_salary = np.median(salary)
```

```
print(median_salary)
```

```
std_salary = np.std(salary)
```

```
print(std_salary)
```

```
total_salary = np.sum(salary)
```

```
print(total_salary)
```

```
sns.histplot(df['Salary'])
```

```
sns.catplot(data=df,col = "Name",x = "Salary",kind = "count")
```

```
df_h = df.groupby(['Name' , 'Salary']).size().head(10)
```

```
(df_h.head())
```

```
df_h.plot(kind='bar')
plt.title('total salary')
plt.xlabel('Name')
plt.ylabel('Highest salary')
```

```
top_5 = df_h.sort_values(ascending = True).head(5)
```

```
top_5.plot(kind='bar')
plt.title('total salary')
plt.xlabel('Name')
plt.ylabel('Highest salary')
```

```
df_h = df.groupby([ 'Department']).size().head(10)
```

```
print(df_h)
```

```
df_h.plot(kind = 'pie')
plt.title('total salary')
plt.xlabel('Name')
plt.ylabel('Highest salary')
```

```
df_h.plot(kind = 'line')
plt.title('total salary')
plt.show()
```

```
age_info = df[df['Age'].notnull()][['Age']]
```

```
age_info_clean = age_info[age_info['Age'] != 0]
```

```
age_info_clean.info()
```

```
fig = plt.figure(figsize = (20,8))
Age_plot = sns.distplot(age_info_clean, bins = 30)
Age_plot.tick_params(labelsize = 13)
mean = age_info['Age'].mean()
plt.xlabel('Age', fontdict = {'fontsize':15})
plt.title('Age of disease-affected humans', fontdict = {'fontsize':15})
plt.bold=True
plt.axvline(mean,color = 'red')
plt.show()
```
